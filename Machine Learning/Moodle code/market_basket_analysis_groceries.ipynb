用于识别频繁项集，是关联规则学习中的一种常用方法；association——rules函数用于从前面使用的apriori中找到频繁项集中生成关联规则

```python
df.columns = ['items'] # 添加表头
df
```

```python
# .values：这个属性用于获取Series中的值
txn_arr = df['items'].values

# 用于获取前10个元素
txn_arr[0:10] # diplaying first 10 transactions in txn_arr
```

```python
# 创建空列表txn_list，用于将来储存分隔后交易记录
txn_list = [] # creating an empty transactions list called txn_list

#遍历list里所有元素；[]里的每个元素用，隔开
for i in range(0,len(txn_arr)):
    txn_list.append(txn_arr[i].split(','))


txn_list[0:10] # displaying first 10 transactions in txn_list
```

```python
# 创建"enc"的TransactionEncoder用于交易数据的编码矩阵
## （通常是Bool）
enc = TransactionEncoder()

# 使用矩阵对txt_list拟合fit&转换transform
## 矩阵每行代表一种商品，如果包含某商品则代表TRUE
txn_list_enc = enc.fit(txn_list).transform(txn_list)

#展示矩阵里的前十个交易
txn_list_enc[0:10]
```

```python
# New DataFrame txn_df_enc;data用的是转置后的矩阵；
##enc.columns_ 应该是由 TransactionEncoder enc提供的，包含所有唯一商品的名称
### dtype是把布尔值转换成0,1
txn_df_enc = pd.DataFrame(data = txn_list_enc,
                          columns = enc.columns_,
                          dtype = int)
txn_df_enc.head(3)
```

* Support of X (Sx) = Nx/N
* Support of Y (Sy) = Ny/N
* Support of (X -> Y) = Nxy/N



1.   使用 apriori 函数从 txn_df_enc DataFrame中生成频繁项；这是市场篮子分析的第一步
2.   通过对这些频繁项集按支持度进行排序，我们可以快速识别出在交易数据中最常一起购买的商品或商品组合



min_support最小支持度：项集在所有交易中出现的频率，具体说就是支持度是项集出现的次数与总交易次数的比例。

```python
#apriori生成频繁项集（列名是商品种类）
#use_colnames=True 参数指示 apriori 函数使用列名而不是列索引来表示项集
## 参数设置最小支持度阈值为20%，即交易出现频率至少高于20%
df_support = apriori(txn_df_enc, use_colnames = True, min_support = 0.2)

# 对df_support用sort进行降序排列
## ascending = False是降序，TRUE时是升序；默认是升序
df_support.sort_values(by = 'support', ascending = False)
```

* Lift (X -> Y) = Nxy/Nx*Ny

```python
#使用association_rules从频繁项集Support中生成关联规则
## metric'lift'表示基于lift的运算作为筛选标准
### 只有lift>=1才会被考虑
df_lift = association_rules(df_support, metric = 'lift', min_threshold = 1)

#对lift进行降序排列
df_lift.sort_values(by = 'lift', ascending = False)
```

Since these items have lift > 1, they are usually complement products

```python
df_lift = association_rules(df_support, metric = 'lift', min_threshold = 0)
df_lift[df_lift['lift'] < 1] # displaying results with lift < 1
```

These items with lift < 1 are usually substitute of one another

```python
df_lift = association_rules(df_support, metric = 'lift', min_threshold = 0)
df_lift[df_lift['lift'] == 1] # displaying results with lift = 1
```

There are no items with lift = 1, it means that.............

* Confidence (X -> Y) = [P(Y | X) / P(X)] = [(Nxy/N) / (Nx/N)] = Nxy/Nx

置信度的含义：
置信度 = 1：表示每次 A 出现时，B 也必然出现。这是一条非常强的规则，意味着 A 和 B 之间存在非常强的正相关关系

置信度 > 0.5（这个阈值可以根据具体情况调整）：表示大多数包含 A 的交易也包含了B。这意味着 A 和 B 之间有较强的正相关性，但不是绝对的。置信度较高的规则通常被认为是强规则，但需要结合其他指标（如支持度和提升度）全面评估

置信度 < 0.5：表示少于一半包含 A 的交易也包含了 B。这样的规则可能不够强，表示 A 和 B 之间的关联性较弱。

置信度很低：如果置信度接近于0，这意味着在 A 出现的交易中，B 几乎不出现。这表明 A 和 B 之间没有明显的关联，或者是负相关关系。

```python
# Confidence也是基于Support生成关联，但是是基于Confidence规则
df_confidence = association_rules(df_support, metric
                                  = 'confidence', min_threshold = 0.3)

df_confidence.sort_values(by = 'confidence', ascending = False)
```
