```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/sample_data/IMA1 Dataset.csv")
df.head()
```

```python
# Considering Genre_2 and Genre_3 have many missing values and they are nominal variables
# We use "0" to deal with misssing value
df[['Genre_2', 'Genre_3']] = df[['Genre_2', 'Genre_3']].fillna(0)
```

```python
# Drop rows with nmissing value in the dataset
df = df.dropna()
# Check null values in dataset
print(df.isnull().sum())
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing
from mlxtend.preprocessing import TransactionEncoder
# OR...from mlxtend.preprocessing import OnehotTransactions
from mlxtend.frequent_patterns import apriori, association_rules
# .values：这个属性用于获取Series中的值
txn_arr = df['genres'].values

# 用于获取前10个元素
txn_arr[0:10] # diplaying first 10 transactions in txn_arr
```

```python
# 使用列表推导式和字符串的 replace 方法来替换 "\n" 为 ","
txn_arr_replaced = [x.replace('\n', ', ') for x in txn_arr]

# 打印替换后的结果
print(txn_arr_replaced)
```

```python
# 创建空列表txn_list，用于将来储存分隔后交易记录
txn_list = [] # creating an empty transactions list called txn_list

#遍历list里所有元素；[]里的每个元素用，隔开
for i in range(0,len(txn_arr_replaced)):
    txn_list.append(txn_arr_replaced[i].split(','))


txn_list[0:10] # displaying first 10 transactions in txn_list
```

```python
# 重新检查和替换 "\n" 为 ", "
txn_arr_replaced = [x.replace('\n', ', ') for x in txn_arr]

# 分割字符串，确保没有遗漏
txn_list = [x.split(', ') for x in txn_arr_replaced]  # 注意这里是根据', '来分割

# 打印并检查txn_list中的前几项，确认没有带"\n"的字符串
print(txn_list[:10])

# 如果上述步骤正确无误，继续使用 TransactionEncoder
```

```python
# 创建"enc"的TransactionEncoder用于交易数据的编码矩阵
## （通常是Bool）
enc = TransactionEncoder()

# 使用矩阵对txt_list拟合fit&转换transform
## 矩阵每行代表一种商品，如果包含某商品则代表TRUE
txn_list_enc = enc.fit(txn_list).transform(txn_list)

#展示矩阵里的前十个交易
txn_list_enc[0:10]
```

```python
# New DataFrame txn_df_enc;data用的是转置后的矩阵；
##enc.columns_ 应该是由 TransactionEncoder enc提供的，包含所有唯一商品的名称
### dtype是把布尔值转换成0,1
txn_df_enc = pd.DataFrame(data = txn_list_enc,
                          columns = enc.columns_,
                          dtype = int)
txn_df_enc.head()
```
