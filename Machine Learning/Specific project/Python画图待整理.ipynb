### 画雷达图的代码

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 定义维度数量
num_vars = len(df['category'])

# 创建角度
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # 闭合图形

# 绘制雷达图
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# 获取评分并添加第一个评分到列表末尾
ratings = df['rating'].tolist()
ratings += ratings[:1]

# 绘制线条
ax.plot(angles, ratings, marker='o', linestyle='-', color='#1E90FF')  # 使用蓝色的颜色代码

# 填充区域
ax.fill(angles, ratings, alpha=0.25, facecolor='#1E90FF')  # 使用相同的蓝色

# 设置雷达图的范围
ax.set_ylim(0, 10)

# 设置角度刻度位置
ax.set_xticks(angles[:-1])

# 设置角度刻度标签
ax.set_xticklabels(df['category'].tolist())

# 设置标题
# 设置在表头 ax.set_title(' Figure5 Accessibility analysis of the Information Power', size=18, color='black', y=1.1)
# 设置在表下面
fig.suptitle('Figure7 Accessibility Analysis of the Information Power', size=18, color='black', verticalalignment='bottom', y=-0.1)

# 显示雷达图
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

# Create a new figure with a specific size
fig, ax = plt.subplots(figsize=(8, 8))

# Add a large circle to represent e-commerce
e_commerce_circle = Circle((0.5, 0.5), 0.4, color='#3498DB', alpha=0.4, label='E-commerce')
ax.add_patch(e_commerce_circle)

# Add two smaller circles inside for 'Sustainable Last-Mile Delivery' and 'Customer Behavior'
#(0.4, 0.5)：这是圆心的坐标，其中 0.4 是 x 轴上的位置，0.5 是 y 轴上的位置
#0.2：这是圆的半径；alpha=0.5：这个参数设置了圆形的透明度。值 0.5 意味着圆形是半透明的，其中 0 是完全透明，1 是完全不透明
sustainable_circle = Circle((0.5, 0.5), 0.3, color='#5DADE2', alpha=0.4, label='Sustainable Last-Mile Delivery')
customer_circle = Circle((0.5, 0.5), 0.15, color='#AED6F1', alpha=0.4, label='Customer Behavior')
ax.add_patch(sustainable_circle)
ax.add_patch(customer_circle)

# 添加圆心
plt.scatter(0.5, 0.5, s=50, c='red')  # s是点的大小，c是点的颜色

# Add labels manually
plt.text(0.5, 0.525, '3 papers', ha='center', va='center')
plt.text(0.28, 0.5, 'Sustainable\nLast-Mile Delivery', ha='center', va='center', fontsize=7.5)
plt.text(0.5, 0.4, 'Customer\nBehavior', ha='center', va='center', fontsize=9.5)
plt.text(0.5, 0.83, 'E-commerce', ha='center', va='center', color='black', fontsize=12)



# Set the limits of the plot
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

# Remove axis and spines
ax.axis('off')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

# Create a new figure with a specific size
fig, ax = plt.subplots(figsize=(8, 8))

# 定义颜色和边框颜色
color_e_commerce = '#2E86C1'
color_sustainable = '#48C9B0'
color_customer = '#F9E79F'
edge_color = 'black'  # 边框颜色

# Add a large circle to represent e-commerce
e_commerce_circle = Circle((0.5, 0.5), 0.4, color=color_e_commerce, alpha=0.4, edgecolor=edge_color, linewidth=1.5, label='E-commerce')
ax.add_patch(e_commerce_circle)

# Add two smaller circles inside for 'Sustainable Last-Mile Delivery' and 'Customer Behavior'
sustainable_circle = Circle((0.5, 0.5), 0.3, color=color_sustainable, alpha=0.4, edgecolor=edge_color, linewidth=1.5, label='Sustainable Last-Mile Delivery')
customer_circle = Circle((0.5, 0.5), 0.15, color=color_customer, alpha=0.4, edgecolor=edge_color, linewidth=1.5, label='Customer Behavior')
ax.add_patch(sustainable_circle)
ax.add_patch(customer_circle)

# 添加圆心
plt.scatter(0.5, 0.5, s=50, c='red', edgecolors=edge_color, zorder=3)  # zorder=3 使红点在最上层

# Add labels manually
plt.text(0.5, 0.525, '3 papers', ha='center', va='center')
plt.text(0.28, 0.5, 'Sustainable\nLast-Mile Delivery', ha='center', va='center', fontsize=7.5)
plt.text(0.5, 0.4, 'Customer\nBehavior', ha='center', va='center', fontsize=9.5)
plt.text(0.5, 0.83, 'E-commerce', ha='center', va='center', color='black', fontsize=12)

# Set the limits of the plot
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

# Remove axis and spines
ax.axis('off')

# Show the plot
plt.show()
```

### Draw Pyramid

```python
import matplotlib.pyplot as plt

# Revised Bloom's Taxonomy 的六个层级
levels = ['Creating', 'Evaluating', 'Analysing', 'Applying', 'Understanding', 'Remembering']
descriptions = [
    'Create something new',
    'Justify a decision',
    'Explore relationships',
    'Use information in new situations',
    'Explain ideas or concepts',
    'Recall information'
]

# 为每个层级分配一个数值，用于确定金字塔的宽度
level_values = range(len(levels), 0, -1)

# 创建一个颜色列表
colors = ['#ff9999', '#ffcc99', '#ffff99', '#ccff99', '#99ff99', '#99ccff']

fig, ax = plt.subplots()

# 画出金字塔的每一层
for i in range(len(levels)):
    ax.barh(i, level_values[i], align='center', color=colors[i], zorder=10-i) # zorder 控制层叠顺序
    ax.text(level_values[i]/2, i, levels[i], ha='center', va='center', color='black')
    ax.text(level_values[i] + 0.5, i, descriptions[i], ha='left', va='center', color='black')

# 设置 y 轴的标签
ax.set_yticks(range(len(levels)))
ax.set_yticklabels([]) # 清除y轴的标签，因为我们已在条形图中添加了文本

# 移除 x 轴和边框
ax.xaxis.set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# 设置标题
plt.title('Revised Bloom\'s Taxonomy')

plt.show()
```
