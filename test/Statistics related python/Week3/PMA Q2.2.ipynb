```python
### 2.2.1
from scipy.stats import poisson
p_x = poisson.pmf(k=2, mu=1.5)
print(p_x)
```

```python
### 2.2.2
P_x_0 = poisson.pmf(k=0, mu=1.5)
P_x_1 = poisson.pmf(k=1, mu=1.5)
P_less_than_2 = P_x_0 + P_x_1
print(P_less_than_2)
```

```python
### 2.2.3
P_x_square = P_x_0 ** 3
print(P_x_square)
```
