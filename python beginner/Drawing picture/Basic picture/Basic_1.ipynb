### Range() Function

range() 函数在 Python 中用于生成一系列数字。它通常用于循环中，以提供一个可迭代的数字序列。range() 函数有三个参数：start、stop 和 step。
start：序列的开始值，默认为 0。
stop：序列的结束值（不包含此值）。
step：两个值之间的步长，默认为 1。
左闭右开

```python
for i in range(5):
    print(i)
# 输出: 0, 1, 2, 3, 4
```

```python
for i in range(2, 6):
    print(i)
# 输出: 2, 3, 4, 5
```

```python
for i in range(1, 10, 2):
    print(i)
# 输出: 1, 3, 5, 7, 9
```

```python
for i in range(10, 0, -2):
    print(i)
# 输出: 10, 8, 6, 4, 2
```

### Break & Continue

```python
for num in range(10):
    if num == 5:
       break    # 永久跳出循环
    print("Number:", num)

for num in range(10):
    if num == 5: 
       continue #跳出一个循环 : 即 5 这一个循环
    print("Number:", num)
```

### 函数参数

```python
def add(a, b):
    return a + b # return后是函数的表达式

result = add(2, 3)
result
```

```python
def greet(name, msg="Hello"): #默认参数是Hello
    print(f"{msg}, {name}!")

greet("Alice")
greet("Bob", "Good morning")
```

使用 *args 和 **kwargs 传递任意数量的参数。

```python
def fun(*args, **kwargs):
    print("Positional arguments:", args)
    print("Keyword arguments:", kwargs)

fun(1, 2, 3, name="Alice", age=30)
```
